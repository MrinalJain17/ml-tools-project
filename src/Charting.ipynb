```python
%load_ext autotime
from tqdm.auto import tqdm
import os, re
from copy import deepcopy

import pandas as pd
import numpy as np

import plotly.express as px
import plotly.io as pio
import plotly.graph_objs as go
pd.options.plotting.backend = "plotly"
pio.renderers.default = "jpeg"
```

```python
res_path = os.path.join('..', 'data', 'results')
```

### Tuning

```python
algo = 'ac'
```

```python
def chart_tuning(algo):
    # import data and building summary table
    temp_df = pd.read_csv(os.path.join(res_path, 'tuning', f'Bibtex_{algo}_rewards.csv'), index_col=0)
    chart_res = temp_df.cumsum().divide(temp_df.index.values.reshape(-1,1)).iloc[-1].rename('Bibtex')
    chart_res = pd.DataFrame(chart_res)
    
    temp_df = pd.read_csv(os.path.join(res_path, 'tuning', f'Delicious_{algo}_rewards.csv'), index_col=0)
    chart_res['Delicious'] = temp_df.cumsum().divide(temp_df.index.values.reshape(-1,1)).iloc[-1]
    
    temp_df = pd.read_csv(os.path.join(res_path, 'tuning', f'Mediamill_{algo}_rewards.csv'), index_col=0)
    chart_res['Mediamill'] = temp_df.cumsum().divide(temp_df.index.values.reshape(-1,1)).iloc[-1]
    
    temp_df = pd.read_csv(os.path.join(res_path, 'tuning', f'MediamillReduced_{algo}_rewards.csv'), index_col=0)
    chart_res['MediamillReduced'] = temp_df.cumsum().divide(temp_df.index.values.reshape(-1,1)).iloc[-1]
    
    # chart summary table
    fig = chart_res.T.plot.bar(barmode='group')
    fig.update_yaxes(title='Cumulative Mean Reward')
    fig.update_xaxes(title='')
    
    return fig
    
    
```
